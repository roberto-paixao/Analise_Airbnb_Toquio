## Obtenção dos Dados

O portal ISPDados é uma página de Dados Abertos do Instituto de Segurança Pública do Rio de Janeiro, onde podemos [acessar as bases de dados](http://www.ispdados.rj.gov.br/estatistica.html) de registros criminais e de atividade policial do estado do Rio de Janeiro.

De acordo com o site, as estatísticas são construídas a partir dos Registros de Ocorrência (RO) lavrados nas delegacias da Secretaria de Estado de Polícia Civil do Rio de Janeiro (SEPOL), além de informações complementares de órgãos específicos da Secretaria de Estado de Polícia Militar do Rio de Janeiro (SEPM). 

As informações sobre os RO são submetidos ao controle de qualidade realizado pela Corregedoria Geral de Polícia (CGPOL) antes de serem consolidados no ISP.

 As estatísticas produzidas baseiam-se na data em que foi confeccionado o Registro de Ocorrência.

### Importando os dados

Usaremos as seguintes bibliotecas:
* `pandas` - Geração e manipulação de DataFrames.
* `seaborn`, `matplotlib` - Visualização de dados.


```
# Importando os módulos necessários
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

figsize(12,10)
plt.rcParams["font.family"] = "sans-serif"
sns.set_style("white")


# Gerando um DataFrame com os dados baixados.
DATA_PATH = "http://www.ispdados.rj.gov.br/Arquivos/DOMensalEstadoDesde1991.csv"


# O arquivo csv está separado por ';'
df = pd.read_csv(DATA_PATH, sep=';')
```

```
# Soma dos valores nulos do dataset
(df.isnull().sum()).sort_values(ascending=False)
```

Apesar de termos campos em branco, o resultado do somátorio de dados ausentes é `0`. O que indica que este campo vazio, ou seja, pode estar preenchido com alguma coisa.

Por termos atributos do tipo `object` o pandas reconhece os valores em brancos como Strings, como se houvesse o caracter "espaço" neste campo, portanto, precisarão ser convertidos para o tipo numérico.

O `Pandas` possui o método `replace()` que nos permite substituir determinados elementos do _DataFrame_ por outros, então substituiremos esses "espaços" pelo valor nulo `NaN` do `numpy`, no qual o `pandas` atribui os valores nulos.

```
# Substituindo os espaços em brancos por valores Nulos 

df.replace(r" ", np.nan, inplace=True)
```

Agora só precisamos converter essas variáveis do tipo `object` para o tipo `float`.

```
# Gerando uma lista com o nome dos atributos do tipo 'object'
features_obj = df.select_dtypes('object').columns

# Transformando esses atributos para o tipo float
df[features_obj] = df[features_obj].astype('float')
```

Se tudo correr de forma correta, a partir de agora poderemos visualizar os dados realmente faltantes do _dataset_.

Checaremos a soma dos valores nulos novamente.

```
(df.isnull().sum()).sort_values(ascending=False)
```

```
# Plotando a porcentagem de dados ausentes

ax, fig = plt.subplots(figsize=(12,12))

missing_percent = df.isnull().mean().sort_values(ascending=False).round(2)*100
missing_percent = missing_percent[missing_percent > 0]
# Configurações


valores = missing_percent.values
indices = missing_percent.index

# Plot
splot = sns.barplot(valores, indices, orient='h', color='DarkOrange');

for p in splot.patches:
    width=p.get_width()

    if width >= 0.1:
        plt.text(p.get_width()-3, p.get_y()+0.55*p.get_height(),
                    '{}%'.format(width),
                    ha='center', va='center', color='white', size=11, fontweight='light')

        
        
fig.set_title('Porcentagem faltante das features\n', loc='left', size=22, color='grey')
plt.xticks([], color='white')


sns.despine(left=True, bottom=True);


```

Os valores do gráfico acima foram arredondados para melhor visualizarmos esses valores.

### Crimes contra o Patrimônio - Veículos

Um dos crimes mais comuns são os relacionados a veículos.

É um problema bastante recorrente nas metrópoles brasileiras e esta seção será dedicada a esse problema. Por termos dados sobre roubo, furto e recuperação de veículos, conseguiremos usar essas informações para nos situarmos sobre o resultado das ações de combate a esse tipo de crime.

Antes de entrarmos em detalhes, devemos ressaltar a diferença entre _furto_ e _roubo_, pois de acordo com o [Tribunal de Justiça do Distrito Federal e dos Territórios](https://www.tjdft.jus.br/institucional/imprensa/campanhas-e-produtos/direito-facil/edicao-semanal/furto-e-roubo), **furto** é a subtração de um patrimônio sem que haja violência, enquanto **roubo** é caracterizado como subtração do patrimônio sob grave ameaça ou violência. 

Veremos os valores para o roubo de veículos.
